```python
movie.dropna(inplace=True)
```

```python
movie.reset_index(inplace=True)
```

```python
del movie['index']
```

```python
movie
```

```python
for i in range(3576):
    movie['actors'][i] = movie['actors'][i].replace("|",",")
```

```python
for i in range(3576):
    movie['genres'][i] = movie['genres'][i].replace("|",",")
```

```python
movie
```

```python
movie['actors']=movie['actors'].apply(lambda x:x.split())
```

```python
movie['genres']=movie['genres'].apply(lambda x:x.split())
```

```python
movie['summary']=movie['summary'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie['data'] = movie['genres']+movie['actors']+movie['summary']
```

```python
main_dataset = movie[['title_x','imdb_id','data','poster_path','wiki_link']]
```

```python
main_dataset
```

```python
main_dataset['data'] = main_dataset['data'].apply(lambda x:" ".join(x))
```

```python
for i in range(3576):
    main_dataset['data'][i] = main_dataset['data'][i].replace("'","")
```

```python
for i in range(3576):
    main_dataset['data'][i] = main_dataset['data'][i].replace(" ",",")
```

```python
main_dataset['data'][0]
```

```python
main_dataset['data']=main_dataset['data'].apply(lambda x:x.lower())
```

```python
import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vectors = cv.fit_transform(main_dataset['data']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
import pickle
pickle.dump(main_dataset,open('movies.pkl','wb'))
pickle.dump(main_dataset.to_dict(),open('movies_dict.pkl','wb'))
pickle.dump(similarity,open('similarity.pkl','wb'))
movies_dict = pickle.load(open('movies_dict.pkl','rb'))
movies = pd.DataFrame(movies_dict)
similarity = pickle.load((open('similarity.pkl','rb')))
```

```python
def recommend(movie):
    movie_index = movies[movies['title_x'] == movie].index[0]
    ditances = similarity[movie_index]
    movies_list = sorted(list(enumerate(ditances)),reverse=True,key=lambda x:x[1])[1:6]

    recommend_movies = []
    recommend_movies_posters = []
    recommend_movies_link = []
    for i in movies_list:
        recommend_movies.append(movies.iloc[i[0]].title_x)
        recommend_movies_posters.append(movies.iloc[i[0]].poster_path)
        recommend_movies_link.append(movies.iloc[i[0]].wiki_link)
    return recommend_movies,recommend_movies_posters,recommend_movies_link
```

```python
recommend('Deewaar')
```
